```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import io
from scipy import stats
```

```python
from google.colab import files
up = files.upload()

```

```python
data = pd.read_csv(io.StringIO(up['Cutlets.csv'].decode('utf-8')))
data
```

```python
data.info()
```

```python
data.isnull()
```

```python
# Assumption  h0 = There is no difference in the size of the cutlets of these 2 groups, diameter A = diameter B
#             h1 = diameter A != diameter B
```

```python
CutletA = data['Unit A']
CutletA
```

```python
CutletB = data['Unit B']
CutletB
```

```python
stats.ttest_ind(CutletA,CutletB)
```

```python
# since pvalue is greater than 0.05, hence assumption h0 is accepted.  
# There is no difference in the size of the cutlets of these 2 groups, diameter A = diameter B.
```

```python
data3.info()
```

```python
# Assumption > h0 = No difference in buyer ratio of male-female. 
#              h1 = There is a difference in buyer ratio of male-female. 
# Applying chi-sq. test
```

```python
male = [50, 142, 131, 70]
female = [435, 1523, 1356, 750]
new_data = [male, female]
new_data
```

```python
stat,p,dof,exp = stats.chi2_contingency(new_data)
```

```python
p
```

```python
# as p is greater than alpha(0.05), we accept h0.
```
